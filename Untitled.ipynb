```python
PATTERN_LINEBEGIN = re.compile(r"([A-Z]+?:)(.*)")
interview_lines = []
for p in paragraphs:
    m = PATTERN_LINEBEGIN.search(p.text)
    if m:
        interview_lines.append((m.group(1), m.group(2).strip()))
```

```python
interview_lines[5]
```

```python
import spacy
```

```python
nlp = spacy.load('en_core_web_sm')
```

```python
# start = timer()
# docs = [nlp(t[1]) for t in interview_lines]
# print(format_timespan(timer()-start))
```

```python
start = timer()
docs = []
for t in interview_lines:
    line = " ".join(t)
    docs.append(nlp(line))
print(format_timespan(timer()-start))
```

```python
for doc in docs:
    for ent in doc.ents:
        print(ent.text, ent.label_)
```

```python
coll = []
for doc in docs:
    ent_texts = [ent.text for ent in doc.ents]
    ent_texts = " ".join(ent_texts)
    if 'Starbucks' in ent_texts:
        coll.append(doc)
```

```python
displacy.render(docs, style='ent', jupyter=True)
```
